```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
adj = pd.read_csv("./running_all_d3.csv", header = 0)
A = adj.to_numpy().transpose()
A
```

```python
place_diff = pd.read_csv("./running_all_d3_diff.csv", header = 0)
L = place_diff.to_numpy().transpose()
L
```

```python
alpha = 0.1
K_in = np.diag(A.sum(axis = 0))
K_out = np.diag(A.sum(axis = 1))
alpha_mat = np.diag(np.repeat(alpha, len(A)))
LHS = K_out + K_in - (A + A.transpose()) + alpha_mat
Z = np.diag(np.diag(A @ L.transpose()))
W = np.diag(np.diag(A.transpose() @ L))
RHS = (Z - W).sum(axis = 1)
s = np.linalg.inv(LHS) @ RHS 
```

```python
rankings = [x for _, x in sorted(zip(s, adj.columns.tolist()),  reverse = True)]
```
